```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from pandas.core.common import random_state
```

```python
#step 1 :dataset load
file  = pd.read_csv(r"C:\Users\ASUS\Downloads\Salary_Data.csv")
file.head()
```

```python
#step 2 :data analysis

file.shape
file['YearsExperience'].count()
file.describe()
```

```python
#step 5 : plot
plt.title('Salary Distribution Plot')
sn.distplot(file['Salary'])
plt.show()
```

```python
#relationship between salary and experience

plt.scatter(file['YearsExperience'],file['Salary'])
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.box(False)
plt.show()
```

```python
#seaborn map

print(file.corr(numeric_only=True))
```

```python
sn.heatmap(file.corr(numeric_only=True),annot=True)
```

```python
#step 4 :Split the dataset into dependent/independent variables
#Experience (X) is the independent variable
#Salary (y) is dependent on experience

X=file.iloc[:,:1]
y=file.iloc[:,1:]

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
#Step 5: Train the regression model

model = LinearRegression()
model.fit(X_train,y_train)
```

```python
#step 6 : find accuracy
accuracy = model.score(X_test,y_test)
print("Accuray is {}".format(round(accuracy * 100)))
```

```python
#predict the output

y_predict= model.predict(X_test)
```

```python
plt.scatter(X_test,y_test)
plt.plot(X_test,y_predict,color='red')
```
